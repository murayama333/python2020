# pandas - part6 - DataFrameのグループ化 - Training

---

```
import pandas as pd
import japanize_matplotlib

pd.options.display.precision = 3
pd.options.display.float_format = '{:.3f}'.format

rank = pd.read_csv("rank.csv", index_col=["チームID", "背番号"])
rank
```

## 6.1 チームIDでグループ化して名前の数（件数）を出力してください。




```
rank.groupby("チームID")[["名前"]].count()
```

## 6.2 チームIDでグループ化して本塁打の最大数を出力してください。本塁打の降順でソートするものとします。



```
rank.groupby("チームID")[["本塁打"]].max().sort_values("本塁打", ascending=False)
```

## 6.3 打数を500以上、500未満に分けて打数グループ列を追加します。打数グループ列でグループ化して、打率の最大値を出力してください。

```
rank["打数グループ"] = "500未満"
rank.loc[rank["打数"] >= 500, ["打数グループ"]] = "500以上"
rank.groupby("打数グループ")[["打率"]].max()
```

# pandas - part7 - DataFrameの結合 - Training

---


```
import pandas as pd
import japanize_matplotlib

pd.options.display.precision = 3
pd.options.display.float_format = '{:.3f}'.format

rank = pd.read_csv("rank.csv", index_col=["チームID", "背番号"])
rank
```

## team.csvのダウンロード

```
!wget https://gist.githubusercontent.com/murayama333/537b2cd02290d542905e2dd6f035694d/raw/a62d467780db34ed976f04797f25a83c50a3ef22/team.csv
```

```
team = pd.read_csv("team.csv", index_col="ID")
team
```

## 7.1 rank と team を左外部結合して結果を出力してください。

```
rank.join(team, on="チームID")
```

## 7.2 rank と team を左外部結合して、名前、チーム名、本拠地、打率、本塁打、OPSを出力してください。OPSの降順でソートするものとします。

```
rank.join(team, on="チームID")[["名前", "チーム名", "本拠地", "打率", "本塁打", "OPS"]].sort_values("OPS", ascending=False)
```

## league.csvのダウンロード

```
!wget https://gist.githubusercontent.com/murayama333/18e91bcccdf7962d5fa28b1bc723a427/raw/da3da50d5db12f85765d35bcc762647818700f2d/league.csv
```

```
league = pd.read_csv("league.csv", index_col="ID")
league
```

## 7.3 rank、team、leagueをそれぞれ左外部結合して、結果を出力してください。

```
rank.join(team, on="チームID").join(league, on="リーグID")
```

## 7.4 rank、team、leagueをそれぞれ左外部結合し、リーグ名が"LEAGUE A"の行について、リーグ名、チーム名、名前、OPSを出力してください。OPSの降順でソートするものとします。

```
rank_team_league = rank.join(team, on="チームID").join(league, on="リーグID")
rank_team_league[rank_team_league["リーグ名"] == "LEAGUE A"][["リーグ名", "チーム名", "名前", "OPS"]].sort_values("OPS", ascending=False)
```

## 7.5 rank、team、leagueをそれぞれ左外部結合し、リーグ名でグループ化して打率と本塁打の最大値を出力してください。

```
rank.join(team, on="チームID").join(league, on="リーグID").groupby("リーグ名")[["打率", "本塁打"]].max()
```

# pandas - part8 - 欠損値の補完 - Training

## rank_na.csvのダウンロード

```
!wget https://gist.githubusercontent.com/murayama333/1f04e8cfedf4327e5649dc770d8d9e82/raw/3de390adfefe0b6731a4a69fba75bca574ad94ef/rank_na.csv
```

```
import pandas as pd
import japanize_matplotlib

pd.options.display.precision = 3
pd.options.display.float_format = '{:.3f}'.format

rank = pd.read_csv("rank_na.csv", index_col=["チームID", "背番号"])
rank
```

> 打数、三塁打にそれぞれ欠損値が存在することを確認します。

## 8.1 欠損値を含む行を削除して結果を出力してください。

```
rank.dropna()
```

## 8.2 欠損値を含む行だけを出力してください。

```
rank[rank.isnull().any(axis=1)]
```

## 8.3 欠損値を１つ前の行から補完して結果を出力してください。

```
rank.fillna(method="ffill")
```

## 8.4 欠損値を最小値で補完して結果を出力してください。

```
rank.fillna(rank.min())
```
